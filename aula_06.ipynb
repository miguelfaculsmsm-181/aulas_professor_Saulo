## Exemplo Resolvido 1: Tabuada

**O problema:** Crie um programa que peça um número ao usuário e imprima a tabuada desse número de 1 a 10 usando o `for`.

```python
numero = int(input("Você quer a tabuada de qual número? "))

print(f"\n--- Tabuada do {numero} ---")
for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} x {i:2d} = {resultado:2d}")
```

*Dica: O formato `{:2d}` dentro da f-string diz para o Python reservar 2 espaços na hora de imprimir aquele número. Isso ajuda a deixar os números alinhados bonitinhos na tela!*

## Exemplo Resolvido 2: Somatória com While

**O problema:** Peça números ao usuário repetidamente. O programa deve somar todos os números digitados e só deve parar (e mostrar o total) quando o usuário digitar o **número ZERO**.

```python
soma = 0
while True:
    valor = float(input("Digite um valor para somar (ou 0 para parar): "))

    if valor == 0:
        break # Sai do loop quando o usuário digitar zero

    soma = soma + valor

print(f"\nA soma total foi de: {soma}")
```

### Desafio 3.
Supondo que a população de um país A seja da ordem de 80000 habitantes com uma taxa anual de crescimento de 3% e que a população de B seja 200000 habitantes com uma taxa de crescimento de 1.5%. Faça um programa que calcule e escreva o número de anos necessários para que a população do país A ultrapasse ou iguale a população do país B, mantidas as taxas de crescimento.

```python
# Escreva seu código aqui

pop_a = 80000
pop_b = 200000
anos = 0

while pop_a < pop_b:
    pop_a = pop_a * 1.03   # crescimento de 3%
    pop_b = pop_b * 1.015  # crescimento de 1,5%
    anos += 1

print(f"Serão necessários {anos} anos.")
print(f"População de A: {pop_a:.0f}")
print(f"População de B: {pop_b:.0f}")
```

### Desafio 4.
Altere o programa anterior permitindo ao usuário informar as populações e as taxas de crescimento iniciais. Valide a entrada e permita repetir a operação.

```python
# Escreva seu código aqui

while True:
    # Entrada e validação
    pop_a = int(input("Digite a população do país A: "))
    while pop_a <= 0:
        pop_a = int(input("Valor inválido. Digite uma população maior que 0: "))

    pop_b = int(input("Digite a população do país B: "))
    while pop_b <= 0:
        pop_b = int(input("Valor inválido. Digite uma população maior que 0: "))

    taxa_a = float(input("Digite a taxa de crescimento de A (%): "))
    while taxa_a <= 0:
        taxa_a = float(input("Valor inválido. Digite uma taxa maior que 0: "))

    taxa_b = float(input("Digite a taxa de crescimento de B (%): "))
    while taxa_b <= 0:
        taxa_b = float(input("Valor inválido. Digite uma taxa maior que 0: "))

    anos = 0

    while pop_a < pop_b:
        pop_a = pop_a * (1 + taxa_a / 100)
        pop_b = pop_b * (1 + taxa_b / 100)
        anos += 1

    print(f"\nSerão necessários {anos} anos para que A alcance ou ultrapasse B.")

    repetir = input("\nDeseja realizar outro cálculo? (s/n): ").lower()

    if repetir != "s":
        print("Programa encerrado.")
        break
```

### Desafio 21.
Faça um programa que peça um número inteiro e determine se ele é ou não um número primo. Um número primo é aquele que é divisível somente por ele mesmo e por 1.

```python
# Escreva seu código aqui

num = int(input("Digite um numero inteiro: "))

if num <= 1:
  print("Não é primo")
else:
  primo = True

  for i in range(2, num):
    if num % i == 0:
      primo = False
      break

  if primo:
    print("É primo")
  else:
    print("Não é primo")
```

### Desafio 23.

```python
# Escreva seu código aqui

n = int(input("Digite Um numero inteiro: "))

toltal_divisoes = 0

for num in range(2, n + 1):
  primo = True
  for i in range(2, num):
    toltal_divisoes += 1
    if num % i == 0:
      primo = False
      break

  if primo:
    print(num)

print("Total de divisões realizadas:", toltal_divisoes)
```

### Desafio 25.
Faça um programa que peça para n pessoas a sua idade, ao final o programa devera verificar se a média de idade da turma varia entre 0 e 25,26 e 60 e maior que 60; e então, dizer se a turma é jovem, adulta ou idosa, conforme a média calculada.

```python
num_pessoas = int(input("Quantas pessoas irão participar da pesquisa? "))

soma_idades = 0
for i in range(num_pessoas):
  idade = int(input(f"Digite a idade da pessoa {i + 1}: "))
  soma_idades += idade

if num_pessoas > 0:
  media_idade = soma_idades / num_pessoas
  print(f"\nMédia de idade da turma: {media_idade:.2f}")

  if 0 <= media_idade <= 25:
    print("A turma é jovem.")
  elif 26 <= media_idade <= 60:
    print("A turma é adulta.")
  elif media_idade > 60:
    print("A turma é idosa.")
else:
  print("Nenhuma idade foi informada para calcular a média.")
```

### Desafio 26.
Numa eleição existem três candidatos. Faça um programa que peça o número total de eleitores. Peça para cada eleitor votar e ao final mostrar o número de votos de cada candidato.

```python
num_eleitores = int(input("Digite o número total de eleitores: "))

votos_candidato1 = 0
votos_candidato2 = 0
votos_candidato3 = 0

for i in range(1, num_eleitores + 1):
  while True:
    voto = int(input(f"Eleitor {i}: Digite seu voto (1, 2 ou 3 para o candidato): "))
    if voto in [1, 2, 3]:
      break
    else:
      print("Voto inválido! Por favor, digite 1, 2 ou 3.")

  if voto == 1:
    votos_candidato1 += 1
  elif voto == 2:
    votos_candidato2 += 1
  elif voto == 3:
    votos_candidato3 += 1

print("\n--- Resultado da Eleição ---")
print(f"Candidato 1: {votos_candidato1} votos")
print(f"Candidato 2: {votos_candidato2} votos")
print(f"Candidato 3: {votos_candidato3} votos")
```

### Desafio 27.
Faça um programa que calcule o número médio de alunos por turma. Para isto, peça a quantidade de turmas e a quantidade de alunos para cada turma. As turmas não podem ter mais de 40 alunos.

```python
num_turmas = int(input("Digite a quantidade de turmas: "))

total_alunos = 0

for i in range(1, num_turmas + 1):
  while True:
    alunos_na_turma = int(input(f"Digite a quantidade de alunos para a turma {i} (máximo 40): "))
    if 0 <= alunos_na_turma <= 40:
      total_alunos += alunos_na_turma
      break
    else:
      print("Quantidade de alunos inválida! A turma não pode ter mais de 40 alunos (e não menos de 0).")

if num_turmas > 0:
  media_alunos_por_turma = total_alunos / num_turmas
  print(f"\nO número médio de alunos por turma é: {media_alunos_por_turma:.2f}")
else:
  print("Nenhuma turma foi informada.")
```
