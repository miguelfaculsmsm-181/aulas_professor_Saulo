# Aula 08: Dicionários

Nós vimos que Listas são ótimas para guardar uma coleção de itens ordenados (índice 0, 1, 2...). Mas e se quisermos guardar as características de um aluno? Como `nome`, `idade`, `nota`?
Para isso, usamos os **Dicionários**. Eles guardam os dados em pares de **Chave e Valor**.

## 1. Criando e Acessando Dicionários

Dicionários são criados usando chaves `{}` (curly braces).

```python
aluno = {
    "nome": "Carlos",
    "idade": 21,
    "curso": "Engenharia Civil",
    "nota": 8.5
}

print("Dicionário completo:", aluno)
```

Em vez de acessar pelo índice numérico `[0]`, nós acessamos o valor usando a sua **Chave**:

```python
print("Nome do aluno:", aluno["nome"])
print("Nota de", aluno["nome"], "foi:", aluno["nota"])
```

## 2. Adicionando e Modificando Valores

É muito simples atualizar ou adicionar novos pares de chave:valor.

```python
# Modificando um valor existente
aluno["nota"] = 9.0
print("Nova nota:", aluno["nota"])

# Adicionando uma nova chave
aluno["aprovado"] = True
print("Aluno com nova chave:", aluno)
```

## 3. Removendo Itens

Podemos usar o comando `del` ou o método `.pop()` para remover itens.

```python
del aluno["curso"] # Remove a chave curso por completo
print("Sem o curso:", aluno)

# O pop() remove a chave e te devolve o valor dela, caso queira guardar
idade_removida = aluno.pop("idade")
print(f"Foram removidos os {idade_removida} anos da chave. Sobrou:", aluno)
```

## 4. Iterando sobre Dicionários

Como percorremos um dicionário com o `for`? Podemos iterar pelas chaves, pelos valores, ou por ambos!

```python
cidade = {
    "nome": "São Paulo",
    "estado": "SP",
    "habitantes": 12000000
}

print("--- Iterando pelas CHAVES e VALORES ---")
# O .items() nos dá a chave e o valor simultaneamente
for chave, valor in cidade.items():
    print(f"A chave {chave} guarda o valor {valor}")
```

### Desafio 1: Tabela de Preços
Crie um dicionário com 3 nomes de produtos e seus respectivos preços (ex: 'maçã': 3.50). Peça ao usuário para digitar o nome de um produto. Se o produto existir no dicionário, imprima seu preço. Se não existir, imprima 'Produto não encontrado'.

```python
# Dica: você pode verificar se uma chave existe usando o operador 'in'
# if nome_produto in dicionario:
produtos = {
    "maçã": 3.50,
    "banana": 2.80,
    "laranja": 4.00
}

# Seu código aqui
produto = input("digite o nome do produto: ").lower()

if produto in produtos:
  print("Preço:", produtos[produto])
else:
  print("Produto não encontrado.")
```

#### Exercício 3: Gerenciamento de Tarefas

#### Exercício 4: Listando Contatos
Crie um dicionário `contatos` onde as chaves são nomes de pessoas e os valores são seus números de telefone. Itere sobre o dicionário e imprima cada nome e seu respectivo telefone no formato "Nome: Telefone".

```python
contatos = {
    "Ana": "9999-8888",
    "Bruno": "7777-6666",
    "Carla": "5555-4444"
}

# Seu código aqui
for nome, telefone in contatos.items():
    print(f"{nome}: {telefone}")
```

#### Exercício 5: Buscando um País
Crie um dicionário `paises_e_capital` com pelo menos 3 países e suas capitais. Peça ao usuário para digitar o nome de um país. Se o país estiver no dicionário, imprima sua capital. Caso contrário, imprima 'País não encontrado no dicionário'.

```python
paises_e_capital = {
    "Brasil": "Brasília",
    "Argentina": "Buenos Aires",
    "Chile": "Santiago",
    "França": "Paris"
}

# Seu código aqui
pesquisa_pais = input("Digite o nome de um país: ").capitalize()

if pesquisa_pais in paises_e_capital:
    print(f"A capital de {pesquisa_pais} é {paises_e_capital[pesquisa_pais]}")
else:
    print("País não encontrado no dicionário.")
```
