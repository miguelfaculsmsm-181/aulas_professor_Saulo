# Aula 09: Funções

Até hoje, nós escrevemos o código de cima para baixo. Mas e se precisarmos repetir um mesmo bloco de código em várias partes do programa (por exemplo, limpar a tela e mostrar um menu)?
As **Funções** nos permitem dar um *nome* para um bloco de código, para que possamos "chamá-lo" quando quisermos, ajudando a organizar nosso programa.

## 1. Criando uma Função

Para "definir" uma função, usamos a palavra `def`, seguida do nome da função e parênteses `()`.

```python
def dar_boas_vindas():
    print("----------------------")
    print(" Bem vindo ao Sistema ")
    print("----------------------")

# A função não faz nada até ser explicitamente "chamada"
print("Iniciando...")
dar_boas_vindas()
print("Fazendo algo no sistema...")
dar_boas_vindas()
```

## 2. Passando Parâmetros (Argumentos)

Funções podem receber dados de fora para trabalhar com eles. Chamamos isso de "parâmetros".

```python
def saudar_pessoa(nome_da_pessoa):
    print(f"Olá, {nome_da_pessoa}! Seja muito bem-vindo.")

saudar_pessoa("Maria")
saudar_pessoa("Pedro")
```

Podemos ter múltiplos parâmetros, separados por vírgula.

```python
def somar(a, b):
    resultado = a + b
    print(f"A soma de {a} com {b} é {resultado}")

somar(10, 5)
somar(100, 200)
```

## 3. Retornando Valores (`return`)

A função `somar` acima imprimiu o valor na tela. Mas e se quiséssemos pegar esse resultado para usar em outra conta? Funções podem **devolver** um resultado para quem as chamou usando a palavra `return`.

```python
def multiplicar(a, b):
    resultado = a * b
    return resultado # "Devolve" o valor, mas não imprime!

# Guardamos o retorno em uma variável
dobro_de_10 = multiplicar(10, 2)
print(f"O retorno da função foi {dobro_de_10}")

# Assim podemos usar o resultado em outra parte do código
print(f"O quádruplo de 10 é {dobro_de_10 * 2}")
```

## 4. Retornando Múltiplos Valores (Tuplas)

Uma característica incrível e muito usada no Python é a capacidade de uma função retornar **mais de um valor ao mesmo tempo**. Diferente de linguagens como C ou Java (onde você normalmente tem que fazer truques ou criar classes para devolver dois valores), no Python basta separá-los por vírgula no `return`.

```python
def calcular_estatisticas(numeros):
    # Calcula três coisas de uma vez
    total = sum(numeros)
    maior = max(numeros)
    menor = min(numeros)

    # Retornamos os três valores separados por vírgula.
    # Isso magicamente os agrupa no que o Python chama de 'Tupla' (parecida com uma lista, mas imutável)
    return total, maior, menor

minhas_notas = [7, 8, 10, 5, 9]

# Na hora de chamar a função, também usamos variáveis separadas por vírgula para receber as 3 respostas!
# Chamamos isso de 'Desempacotamento' (Unpacking)
soma_total, nota_maxima, nota_minima = calcular_estatisticas(minhas_notas)

print(f"Soma: {soma_total}")
print(f"Máxima: {nota_maxima}")
print(f"Mínima: {nota_minima}")
```

Se você guardar a resposta numa variável só (sem desempacotar), o Python te devolverá os itens entre parênteses `()`. Isso é o que define uma **Tupla**.

```python
tudo_junto = calcular_estatisticas(minhas_notas)
print("Retorno direto (Tupla):", tudo_junto)

# Tuplas funcionam igual a listas para acessar pelo índice, mas você não pode alterá-las depois de criadas:
print("Acessando o primeiro item da tupla (soma):", tudo_junto[0])
```

### Desafio 3: Soma de Elementos de uma Lista
Crie uma função `soma_lista` que recebe uma lista de números e retorna a soma de todos os elementos. Teste a função com uma lista de sua escolha.

```python
def soma_lista(lista):
    soma = 0

    for numero in lista:
        soma += numero

    return soma


numeros = [10, 20, 30, 40, 50]

resultado = soma_lista(numeros)

print("A soma dos elementos da lista é:", resultado)
```

### Desafio 4: Contagem de Pares
Crie uma função `contar_pares` que recebe uma lista de números e retorna quantos números pares existem na lista. Use um laço de repetição para iterar sobre a lista.

```python
def contar_pares(lista):
    quantidade = 0

    for numero in lista:
        if numero % 2 == 0:
            quantidade += 1

    return quantidade


numeros = [2, 5, 8, 11, 14, 17, 20]

resultado = contar_pares(numeros)

print("Quantidade de números pares:", resultado)
```

### Desafio 5: Encontrar o Maior Número
Escreva uma função `encontrar_maior` que recebe uma lista de números e retorna o maior número presente na lista. Não use a função `max()` embutida do Python; implemente a lógica com um laço.

```python
def encontrar_maior(lista):
    maior = lista[0]

    for numero in lista:
        if numero > maior:
            maior = numero

    return maior


numeros = [12, 45, 7, 89, 34, 56]

resultado = encontrar_maior(numeros)

print("O maior número da lista é:", resultado)
```

### Desafio 6: Filtrar Números Maiores que X
Crie uma função `filtrar_numeros` que recebe uma lista de números e um valor `x`. A função deve retornar uma **nova lista** contendo apenas os números da lista original que são maiores que `x`.

```python
def filtrar_numeros(lista, x):
    nova_lista = []

    for numero in lista:
        if numero > x:
            nova_lista.append(numero)

    return nova_lista


numeros = [5, 12, 8, 20, 3, 15]
resultado = filtrar_numeros(numeros, 10)

print("Números maiores que 10:", resultado)
```

### Desafio 7: Inverter Lista
Crie uma função `inverter_lista` que recebe uma lista e retorna uma nova lista com os elementos na ordem inversa. Não use `list.reverse()` ou `[::-1]`.

```python
def inverter_lista(lista):
    lista_invertida = []

    for i in range(len(lista) - 1, -1, -1):
        lista_invertida.append(lista[i])

    return lista_invertida


numeros = [1, 2, 3, 4, 5]

resultado = inverter_lista(numeros)

print("Lista original:", numeros)
print("Lista invertida:", resultado)
```

### Desafio 9: Verificar se Elemento Existe
Crie uma função `existe_elemento` que recebe uma lista e um elemento. A função deve retornar `True` se o elemento estiver na lista e `False` caso contrário. Use um laço de repetição.

```python
def existe_elemento(lista, elemento):
    for item in lista:
        if item == elemento:
            return True

    return False


numeros = [10, 20, 30, 40, 50]

print(existe_elemento(numeros, 30))
print(existe_elemento(numeros, 25))
```

### Desafio 10: Média de Números Positivos
Escreva uma função `media_positivos` que recebe uma lista de números. A função deve calcular e retornar a média apenas dos números positivos na lista. Se não houver números positivos, retorne `0`.

```python
def media_positivos(lista):
    soma = 0
    quantidade = 0

    for numero in lista:
        if numero > 0:
            soma += numero
            quantidade += 1

    if quantidade == 0:
        return 0

    return soma / quantidade


numeros = [10, -5, 20, 0, 15, -2]

resultado = media_positivos(numeros)

print("Média dos números positivos:", resultado)
```

### Desafio 11: Lista de Quadrados
Crie uma função `lista_de_quadrados` que recebe um número inteiro `n` e retorna uma lista contendo o quadrado de todos os números de 1 até `n`.

```python
def lista_de_quadrados(n):
    quadrados = []

    for i in range(1, n + 1):
        quadrados.append(i ** 2)

    return quadrados


resultado = lista_de_quadrados(5)

print(resultado)
```

### Desafio 12: Contar Vogais em uma String
Crie uma função `contar_vogais` que recebe uma string como parâmetro e retorna o número total de vogais (a, e, i, o, u, maiúsculas ou minúsculas) nela. Use um laço `for`.

```python
def contar_vogais(texto):
    quantidade = 0

    for letra in texto:
        if letra.lower() in "aeiou":
            quantidade += 1

    return quantidade


frase = "Programacao em Python"

resultado = contar_vogais(frase)

print("Quantidade de vogais:", resultado)
```
